```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
bacnco_color = '#023047'
bancco_color = '#219ebc'
bancoc_color = '#ffb703'
bcanco_color = '#c44900'
conabc_color = '#c9aef1'
nabcco_color = '#ff00bf'
nabcoc_color = '#005c99'
nacobc_color = '#993399'

bacnco_color_light = '#8098a3'
bancco_color_light = '#90cede'
bancoc_color_light = '#ffdb81'
bcanco_color_light = '#e2a480'
conabc_color_light = '#c9aef1'
nabcco_color_light = '#ff80df'
nabcoc_color_light = '#80aecc'
nacobc_color_light = '#cc99cc'
```

```python
small_seq = [18.398, 19.620, 21.206, 22.644, 23.359, 24.125]
medium_seq = [2626.843, 2862.708, 3567.521, 4393.747, 4507.252, 4499.773]
large_seq = [26142.265, 27193.015, 33132.527, 40093.035, 39535.139, 35610.080]

small_hash = [2.608, 5.822, 11.452, 18.579, 22.116, 22.803]
medium_hash = [359.893, 958.334, 2120.016, 4034.014, 4177.741, 4596.419]
large_hash = [3936.377, 8194.25, 17468.769, 30129.032, 34358.803, 34133.509]
x = [10, 25, 50, 75, 90, 100]


# Create a plot
plt.figure()
plt.plot(x, small_seq, marker='o', linestyle='', color=bacnco_color, label='1000 Tuples (Seq)')
plt.plot(x, small_hash, marker='o', linestyle='', color=bancco_color_light, label='1000 Tuples (Index)')
plt.plot(x, medium_seq, marker='o', linestyle='', color=bancoc_color, label='100,000 Tuples (Seq)')
plt.plot(x, medium_hash, marker='o', linestyle='', color=bancoc_color_light, label='100,00 Tuples (Index)')
plt.plot(x, large_seq, marker='o', linestyle='', color=nabcco_color, label='1,000,000 Tuples (Seq)')
plt.plot(x, large_hash, marker='o', linestyle='', color=nabcco_color_light, label='1,000,000 Tuples (Index)')

# plt.xscale('log')
plt.yscale('log')

# Add titles and labels
# plt.title('Execution time of different join orders')
plt.ylabel('Time (in µs)', fontsize=12)
plt.xlabel('Matching Tuples (in percent)', fontsize=12)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize = 12)

# Save the plot as an SVG file
plt.savefig('access.svg', format='svg', bbox_inches='tight')
plt.savefig('access.pdf', format='pdf', bbox_inches='tight')

# Display the plot (optional)
plt.show()
```
